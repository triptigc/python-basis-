## Filtering Outliers (user ages)

```python
raw_ages = [23, -5, 45, 18, 150, 32, -99, 29, 61]

clean_age=[]

for valid_age in raw_ages:
  if valid_age<=0 and valid_age<=110:
    clean_age.append(valid_age)

print(clean_age)
```
